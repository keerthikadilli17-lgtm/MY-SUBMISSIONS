**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')



# Display the first few rows of the data
df.head()


```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)







# Close the connection
conn.close()

```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)


```

## Demo: Running an SQL Query


Count the number of rows in the table named 'main'


```python
QUERY = """
SELECT COUNT(*) 
FROM main
"""

df = pd.read_sql_query(QUERY, conn)
df.head()

```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)


```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age

"""
pd.read_sql_query(QUERY, conn)

```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)



df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
QUERY="SELECT CompTotal FROM main"
df=pd.read_sql_query(QUERY,conn)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df["CompTotal"].dropna(),bins=30)
plt.title("Total Compensation")
plt.xlabel("Comp total")
plt.ylabel("Frequency")

plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
QUERY="SELECT Age,WorkExp FROM main"
df=pd.read_sql_query(QUERY,conn)

```

```python
age_map = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    
    "25-34 years old": 29.5,
    "35-44 years old": 39.5,
    "45-54 years old": 49.5,
    "55-64 years old": 59.5,
    "65 years or older": 70,
    "Prefer not to say": None
}

df["Age_Num"] = df["Age"].map(age_map)

import matplotlib.pyplot as plt


plt.boxplot(df["Age_Num"].dropna())
plt.title("Box Plot of Age")
plt.ylabel("Age")
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
QUERY="SELECT DatabaseWantToWorkWith FROM main"
df=pd.read_sql_query(QUERY,conn)
top5=(df["DatabaseWantToWorkWith"].dropna().value_counts().head(5))
plt.figure(figsize=(8,5))
plt.pie(top5,labels=top5.index,autopct='%1.1f%%',startangle=90)
plt.title("Top 5 Databases Respondents Want to Work With")
plt.axis("equal")
plt.show()




```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here
QUERY="SELECT TimeSearching,TimeAnswering,Age FROM main"
df=pd.read_sql_query(QUERY,conn)
df = df[df["Age"] == "25-34 years old"]
median_values = df[["TimeSearching", "TimeAnswering"]].median()


```

```python
time_map = {
    "Less than 15 minutes a day": 1,
    "15-30 minutes a day": 2,
    "30-60 minutes a day": 3,
    "60-120 minutes a day": 4,
    "Over 120 minutes a day": 5
}

df["TimeSearching_Num"] = df["TimeSearching"].map(time_map)
df["TimeAnswering_Num"] = df["TimeAnswering"].map(time_map)

median_values = df[["TimeSearching_Num", "TimeAnswering_Num"]].median()
```

```python
plt.figure(figsize=(6,5))

median_values.to_frame().T.plot(
    kind="bar",
    stacked=True
)

plt.title("Median TimeSearching and TimeAnswering (25–34 years)")
plt.xlabel("Age Group")
plt.ylabel("Median Time")
plt.legend(["TimeSearching", "TimeAnswering"])

plt.show()
```


### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
QUERY = "SELECT Age, CompTotal FROM main"
df = pd.read_sql_query(QUERY, conn)

df = df[df["Age"].isin(["45-54 years old", "55-64 years old"])]

median_values = df.groupby("Age")["CompTotal"].median()

plt.figure(figsize=(8,5))
plt.plot(median_values.index, median_values.values, marker="o")
plt.title("Median CompTotal (45–60 Age Group)")
plt.xlabel("Age Group")
plt.ylabel("Median CompTotal")
plt.grid(True)
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
QUERY = "SELECT MainBranch FROM main"
df = pd.read_sql_query(QUERY, conn)

counts = df["MainBranch"].value_counts()

plt.figure(figsize=(8,5))
plt.barh(counts.index, counts.values)


plt.title("MainBranch Distribution")
plt.xlabel("Count")
plt.ylabel("MainBranch")

plt.show()
```
