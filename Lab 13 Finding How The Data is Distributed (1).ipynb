### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()

```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Try to find missing columns of entire dataframe
missing_values=df.isnull().sum()
print(missing_values[missing_values>0])

```

```python
##Focusing on Handling missing data in key columns
key_columns=["Employment","JobSat","YearsCodePro"]
print(df[key_columns].isnull().sum())

```

```python
#checking dtypes
df["JobSat"].dtype

```

```python
df["JobSat"]=df["JobSat"].fillna(df["JobSat"].median())
```

```python
df["YearsCodePro"].dtype
```

```python
df["YearsCodePro"]=df["YearsCodePro"].fillna(df["YearsCodePro"]
                                             .mode()[0])

```

```python
df["YearsCodePro"].isnull().sum()

```

```python
df["YearsCodePro"]=df["YearsCodePro"].astype("int")
```

```python
df["YearsCodePro"]=df["YearsCodePro"].replace({"Less than 1 year":"0","More than 50 years":21})
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
sns.kdeplot(df["JobSat"],fill=True)
plt.title("Distribution of Job Satisfaction")
plt.xlabel("JobSatisficaton")
plt.ylabel("Density")
plt.show()
```

```python
##INSIGHTS-
##The Job Satisfication distribution is centered around 7,which reepresents the highest-frequency response.This Sugests that the majority of respondents perceive their job satisfication to be relatively positive.
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
have=df["LanguageHaveWorkedWith"].str.split(':').explode().value_counts()
want=df["LanguageWantToWorkWith"].str.split(':').explode().value_counts()
```

```python
%pip install matplotlib-venn
```

```python

from matplotlib_venn import venn2
import matplotlib.pyplot as plt


have_set=set(have.index)
want_set=set(want.index)

venn2([have_set,want_set],set_labels=("Worked With","Want to Work with"))

plt.title("Programming Language Overlap")
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
remote_trends=pd.crosstab(df["Country"],df["RemoteWork"])
```

```python
from matplotlib.colors import LogNorm
```

```python
plt.figure(figsize=(12,30))

sns.heatmap(remote_trends,cmap="Greens",norm=LogNorm(),annot=False)
plt.title("Remote Work Distribution by Country")
plt.xlabel("Remote Work")
plt.ylabel("Country")

plt.show()
```
